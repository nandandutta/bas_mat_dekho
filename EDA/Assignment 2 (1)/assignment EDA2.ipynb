```python
def color_negative_red(value):
    if value < 0:
        color = 'green'  
    elif value > 0:
        color = 'yellow'
    elif value=="NaN":
        color = 'red'
    elif value==str:
        color = 'green'
    else:
        color = 'red'
        
    return 'color: %s' % color
```
