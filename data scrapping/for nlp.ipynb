```python
import pandas as pd
from bs4 import BeautifulSoup
from tqdm import tqdm
from selenium import webdriver
import requests
import os
```

```python
url='https://in.bookmyshow.com/bengaluru/movies/bala/ET00102272/user-reviews'
```

```python
driver=webdriver.Firefox()
driver.get(url)
```

```python
import time
flag=True
while flag:
    state={}
    zreds=driver.find_elements_by_class_name("zred")
    for zred in zreds:
        if zred.text.strip().lower()=='load more':
            zred.click()
            state[zred]=1
            time.sleep(2)
        else:
            state[zred]=0
    if sum(state.values())==1:
        flag=True
    else:
        flag=False
```

```python
html=driver.page_source
#print(html)
```

```python
html_parsable=BeautifulSoup(html,'html.parser')
```

```python
User_Name=[]
for i in html_parsable.find_all('div',class_='__reviewer-face'):
    User_Name.append(i.text.strip())
#print(User_Name)
len(User_Name)
```

```python
reviews=[]
for i in html_parsable.find_all('div',class_='__reviewer-text'):
    reviews.append(i.text.strip())
#print(reviews)
len(reviews)
```

```python
date=[]
for i in html_parsable.find_all('span',class_='__timestamp'):
    date.append(i.text.strip())
len(date)
```

```python
like=[]
for i in html_parsable.find_all('div',class_='like-dislike _like upvote'):
    like.append(i.text.strip())
len(like)
```

```python
rate=[]
for i in html_parsable.find_all('span',class_='rated-review'):
    rate.append(i['data-value'].strip())
len(rate)
```

```python
s1 =pd.Series(User_Name)
s2=pd.Series(date)
s3=pd.Series(reviews)
s4=pd.Series(like)
s5=pd.Series(rate)
df=pd.DataFrame({'Name':s1,'Date':s2,'Like':s4,'Rate':s5,'Reviews':s3})
```

```python
df.to_csv('Pati_Patni_Aur_Woh.csv')
```
