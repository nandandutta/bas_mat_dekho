```python
base_url="https://www.goodreads.com/quotes?page={}"
```

```python
base_url.format(1)
```

```python
from tqdm import tqdm
authors=[]
qoutes=[]
for i in tqdm(range(1,100)):
    url=base_url.format(i)
    markup=requests.get(url,headers=headers).text
    markup_parsed=BeautifulSoup(markup,'html.parser')
    for i in markup_parsed.find_all('span',class_='authorOrTitle'):
        authors.append(i.text.strip())
    for i in markup_parsed.find_all('div',class_='quoteText'):
        child=list(i.children)[0].strip()
        quotes.append(child)
```

```python
tables=pd.DataFrame({'authors':authors,'quotes':qoutes})
tables.head()
```

```python
base_url1='https://www.goodreads.com/list/show/19.Best%20for%20Book%20Clubs?page={}'
```

```python
from tqdm import tqdm
Bookname=[]
Authors=[]
for i in tqdm(range(1,100)):
    url1=base_url1.format(i)
    markup1=requests.get(url1,headers=headers).text
    markup1_parsed=BeautifulSoup(markup1,'html.parser')
    for i in markup1_parsed.find_all('span',itemprop='author'):
        Authors.append(i.text.strip())
    for i in markup1_parsed.find_all('a',class_='author'):
        Bookname.append(i.text.strip())
```

```python
print(len(Authors))
```

```python
from tqdm import tqdm
base_url = 'https://www.goodreads.com/list/show/19.Best_for_Book_Clubs?page={}'
bookName=[]
authorName=[]
for i in tqdm(range(1,11)):
    url=base_url.format(i)
    markup = requests.get(url,headers=headers).text
    markup_parsed= BeautifulSoup(markup, 'html.parser')
    for i in markup_parsed.find_all('a', class_='bookTitle'):
        bookName.append(i.text.strip())
    for i in markup_parsed.find_all('span', itemprop="author"):
        authorName.append(i.text.strip())
```

```python
print(len(authorName))
```

```python
table=pd.DataFrame({'Book_Name':bookName,'Author_Name': authorName})
table.head()
```
