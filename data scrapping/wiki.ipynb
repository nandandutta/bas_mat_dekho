```python
import pandas as pd
from bs4 import BeautifulSoup
from tqdm import tqdm
from selenium import webdriver
import requests
import os
```

```python
url='https://en.wikipedia.org/wiki/List_of_sandwiches'
driver=webdriver.Firefox()
driver.get(url)
```

```python
html=driver.page_source
#print(html)
```

```python
html_parsable=BeautifulSoup(html,'html.parser')
```

```python
tables=(html_parsable.find_all("table",class_='wikitable'))
```

```python
Name=[]
Origin=[]
Descript=[]
for row in tables[0].find_all("tr")[1:]:
    data=row.find_all("td")
    Name.append(data[0].text.strip())
    Origin.append(data[2].text.strip())
    Descript.append(data[3].text.strip())
Sandwich=pd.DataFrame()
Sandwich['Name']=Name
Sandwich['Origin']=Origin
Sandwich['Descript']=Descript
```

```python
Rank=[]
Film=[]
Year=[]
Director=[]
Studio=[]
Worldwide_gross=[]
for row in tables[2].find_all("tr")[1:]:
    data=row.find_all("td")
    Rank.append(data[0].text.strip())
    Film.append(data[1].text.strip())
    Year.append(data[2].text.strip())
    Director.append(data[3].text.strip())
    Studio.append(data[4].text.strip())
    Worldwide_gross.append(data[5].text.strip())
Bengali=pd.DataFrame()
Bengali['Rank']=Rank
Bengali['Film']=Film
Bengali['Year']=Year
Bengali['Director']=Director
Bengali['Studio']=Studio
Bengali['Worldwide_gross']=Worldwide_gross
```
