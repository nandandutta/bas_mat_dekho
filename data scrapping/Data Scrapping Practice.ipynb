```python
url='https://in.bookmyshow.com/bengaluru/movies'
```

```python
driver=webdriver.Firefox()
driver.get(url)
```

```python
import time
flag=True
while flag:
    state={}
    zreds=driver.find_elements_by_class_name("zred")
    for zred in zreds:
        if zred.text.strip().lower()=='load more':
            zred.click()
            state[zred]=1
            time.sleep(2)
        else:
            state[zred]=0
    if sum(state.values())==1:
        flag=True
    else:
        flag=False
```

```python
html=driver.page_source
print(html)
```

```python
html_parsable=BeautifulSoup(html,'html.parser')
```

```python
User_Name=[]
for i in html_parsable.find_all('div',class_='__reviewer-name'):
    User_Name.append(i.text.strip())
print(User_Name)
len(User_Name)
```

```python
Ratings=[]
for i in html_parsable.find_all('div', class_='__percentage'):
    Ratings.append(i.text.strip())
print(Ratings)
len(Ratings)
```

```python
import re
rat=[]
for i in Ratings:
    rat.append(int(i[:2]))
rat
```

```python
Votes=[]
for i in html_parsable.find_all('div',class_='__count'):
    Votes.append(i.text.strip())
print(Votes)
len(Votes)
```

```python
#converting string value to int
import re
x=[]
y=[]
for i in Votes:
    x.append(i.replace(',','').split()[0])
for j in x:
    if re.search('[0-9]',j):
        y.append(int(j))
    else:
        y.append(0)
y
```

```python
Censors=[]
Language=[]
for i in html_parsable.find_all('div',class_='card-tag'):
    d=i.text.strip().split(' ')
    Language.append(d[2:])
    Censors.append(d[0])
```
